# Training Pipeline

```python
import torch
import random
import torchvision.transforms as transforms
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader, random_split, Subset
import timm
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
transform = transforms.Compose([
    transforms.Resize((128,128)),
    transforms.Lambda(lambda img: img.convert("RGB")),
    transforms.ToTensor(),
])
```

```python
dataset = ImageFolder('/content/khana', transform=transform)
```

```python
random.seed(42)

indices = list(range(len(dataset)))
random.shuffle(indices)

subset_size = int(0.3 * len(dataset))
subset_indices = indices[:subset_size]

dataset_small = Subset(dataset, subset_indices)
```

```python
train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size

train_data, val_data = random_split(dataset, [train_size, val_size])
```

```python
train_loader = DataLoader(train_data, batch_size=64, shuffle=True, num_workers=4)
val_loader = DataLoader(val_data, batch_size=64, num_workers=2)
```

```python
model = timm.create_model('efficientnet_b0', pretrained=True, num_classes=80)
model = model.to(device)
```

```python
from collections import Counter
import torch

labels = [label for _, label in dataset.samples]
class_counts = Counter(labels)

num_classes = len(dataset.classes)

weights = torch.zeros(num_classes)

for i in range(num_classes):
    weights[i] = 1.0 / class_counts[i]

weights = weights.to(device)
```

```python
criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=5e-5, weight_decay=1e-4)
```

```python
print("Starting training...\n")

best_acc = 0

for epoch in range(10):   # fast run
    model.train()

    train_correct, train_total = 0, 0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        outputs = model(images)
        loss = criterion(outputs, labels)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        _, preds = torch.max(outputs, 1)
        train_correct += (preds == labels).sum().item()
        train_total += labels.size(0)

    train_acc = train_correct / train_total

    # validation
    model.eval()
    val_correct, val_total = 0, 0

    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)

            _, preds = torch.max(outputs, 1)
            val_correct += (preds == labels).sum().item()
            val_total += labels.size(0)

    val_acc = val_correct / val_total

    if val_acc > best_acc:
        best_acc = val_acc
        torch.save(model.state_dict(), "best_model.pth")

    print(f"Epoch {epoch+1} done | Train Acc: {train_acc:.4f} | Val Acc: {val_acc:.4f}")

print("\nyipppeeee Training complete!")
print(f"Best Val Accuracy: {best_acc:.4f}")
```
